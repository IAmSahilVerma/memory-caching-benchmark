```python
import importlib

deps = [
    "numpy",
    "einops",
    "tqdm",
    "datasets",
    "tokenizers",
    "torchmetrics",
    "pandas"
]

all_ok = True
for dep in deps:
    try:
        m = importlib.import_module(dep)
        version = getattr(m, "__version__", "unknown")
        print(f" OK {dep:<20} {version}")
    except ImportError:
        print(f" MISSING {dep}")
        all_ok = False
        
print()
print("All dependencies present." if all_ok else "Install missing deps then re-run.")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

a = torch.randn(512, 512, device=device)
b = torch.randn(512, 512, device=device)
c = a @ b

print(f"Matrix multiply on {device}: OK - output shape {c.shape}")
```

```python
from einops import rearrange

x = torch.randn(4, 8, 256)
x_flat = rearrange(x, 'b s h -> (b s) h')
x_back = rearrange(x_flat, '(b s) h -> b s h', b=4)

assert x_back.shape == x.shape
print(f"einops rearrange: OK - {x.shape} -> flat -> {x.shape}")
```
